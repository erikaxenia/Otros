```python
balances_res_pivot=balances_res_pivot.replace(to_replace=np.nan, value=0)
```

```python
multi_index=pd.MultiIndex(levels=[list(balances_res_pivot.columns.levels[0]), list(balances_res_pivot.columns.levels[1])],
                                    labels=list(balances_res_pivot.columns.labels),
                                    names=['variable', 'periodo'])


balances_trim=pd.DataFrame(index=balances_res_pivot.index , columns=multi_index)
```

```python
def trim(row,col,año):
    if balances_trim.loc[balances_trim.iloc[row].name[0]][col][año].name.split('-')[1]=='3':
        
        return balances_pivot.loc[balances_trim.iloc[row].name[0],balances_trim.iloc[row].name[1]][col][año]
    
    else:
        
        return balances_pivot.loc[balances_trim.iloc[row].name[0],balances_trim.iloc[row].name[1]][col][año]-balances_pivot.loc[balances_trim.iloc[row].name[0],balances_trim.iloc[row].name[1]][col][balances_trim.columns.levels[1][list(balances_trim.columns.levels[1]).index(año)-1]]
            
```

```python
# Defino el bucle for al que voy a aplicar la función

variables=['importe']

for j in range(len(balances_trim.index)):
    for col in variables:
        for año in list(balances_trim.columns.levels[1]):
            balances_trim.iloc[j][col][año]=trim(j,col,año)
```
