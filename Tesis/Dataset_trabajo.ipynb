```python
#Filtro las columnas

colum = ['Segmento', 'Cod Cia', 'DenominacionCorta', 'periodo','ind_01','ind_02','ind_03','ind_04','ind_05','ind_06','ind_08','ind_09','ind_10','ind_11','ind_12',
'ind_13','ind_14','ind_15','ind_16','ind_17','ind_18','ind_19','ind_20','ind_21','ind_22','ind_23','ind_24','ind_25',
'ind_26','ind_27','ind_28','ind_29','ind_30','ind_31','ind_32','ind_33','ind_34','ind_35','ind_36','ind_37','ind_38',
'ind_39','ind_40','ind_41','ind_42','ind_43']

data = data[colum]
```

```python
# Reemplazo NA
data = data.fillna(0)

# Remplazo los "den_cero"
data = data.replace(to_replace='den_cero', value=0)
```

```python
# Me quedo a parte con la columna de compañias
encabezado = data[['Segmento','Cod Cia', 'DenominacionCorta','periodo']]

# Filtro todo lo que no sea indicadores
col_num= ['Cod Cia','ind_01','ind_02','ind_03','ind_04','ind_05','ind_06','ind_08','ind_09','ind_10','ind_11','ind_12',
'ind_13','ind_14','ind_15','ind_16','ind_17','ind_18','ind_19','ind_20','ind_21','ind_22','ind_23','ind_24','ind_25',
'ind_26','ind_27','ind_28','ind_29','ind_30','ind_31','ind_32','ind_33','ind_34','ind_35','ind_36','ind_37','ind_38',
'ind_39','ind_40','ind_41','ind_42','ind_43']

indicadores= data[col_num]
```

```python
# Reemplazo punto por coma y lo convierto en flotante

for i in indicadores.columns:
    indicadores[i] = indicadores[i].map(lambda x: x if isinstance(x,int) else float(x.replace(",",".")))
```

```python
indicadores.head()
```

```python
# Junto Dataset

dt_indicadores = encabezado.merge(indicadores, on='Cod Cia', left_index=True, right_index=True)

#dt_indicadores= pd.concat([encabezado, indicadores])
```

```python
dt_indicadores = dt_indicadores.rename(columns = {'Cod Cia':'cod_cia'})
```
